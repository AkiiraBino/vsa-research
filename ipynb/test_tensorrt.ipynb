```python
import ultralytics
```

```python
model = ultralytics.YOLO("/home/akiirabino/video-anal/ipynb/model/yolov8l-d3-e100-default-best.engine")
```

```python
import os
import cv2
root_dir = "/home/akiirabino/video-anal/ipynb/dataset-1/train/images/"
images = [cv2.resize(cv2.imread(root_dir + path), (640,640)) for i, path in enumerate(os.listdir(root_dir)) if i < 100]
```

```python
len(images)
```

```python
result = []
for i in images:
    result.append(model(i))
```

```python
import numpy as np
time = []
for i in result:
    time.append(i[0].speed["inference"])
np.mean(time)
```
